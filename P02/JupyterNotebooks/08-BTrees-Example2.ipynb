### Deleting 
Deleting is trickier because of unimplemented functionality. Finding deletions that will work for both trees is difficult.

```python
def deletetuple(deleteKey):
    results1 = index1.searchByKey(deleteKey)
    results2 = index2.searchByKey(deleteKey)
    rel1.deleteTuple(results1[0])
    rel2.deleteTuple(results2[0])
deletetuple("Srinivasan")
DisplayTwoBTrees("Order=3", index1, "Order=5", index2)
```

```python
deletetuple("Brandt")
DisplayTwoBTrees("Order=3", index1, "Order=5", index2)
```
